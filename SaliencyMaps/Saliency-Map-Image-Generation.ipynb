# Download of GTSRB training and test datasets

```python
import os
import urllib.request 
    
```

```python
def download_gtsrb_training_dataset(folder, filename):

    # Create input folder, if it doesn't exist
    if not os.path.exists(folder):
        os.mkdir(folder)
    
    zip_filepath = os.path.join(folder, filename)
    
    # Download GTSRB training dataset

    print('BEGIN of Download: GTSRB_Final_Training_Images dataset.')

    url = 'http://benchmark.ini.rub.de/Dataset/GTSRB_Final_Training_Images.zip'  
    urllib.request.urlretrieve(url, zip_filepath)

    print('END of Download: GTSRB_Final_Training_Images dataset.\n')
    
    return folder, filename

    
    
def download_gtsrb_test_dataset(folder, filename):

    # Create folder, if it doesn't exist
    if not os.path.exists(folder):
        os.mkdir(folder)
        
    zip_filepath = os.path.join(folder, filename)
    
    # Download GTSRB test dataset

    print('BEGIN of Download: GTSRB_Final_Test_Images dataset.')

    url = 'http://benchmark.ini.rub.de/Dataset/GTSRB_Final_Test_Images.zip'  
    urllib.request.urlretrieve(url, zip_filepath)

    print('END of Download: GTSRB_Final_Test_Images dataset.\n')

    return folder, filename
```

```python
# Download GTSRB Training Dataset from Web
training_dataset_folder, training_dataset_zip_file = download_gtsrb_training_dataset('dataset', 'GTSRB_Training.zip')

# Download GTSRB Test Dataset from Web
test_dataset_folder, test_dataset_zip_file = download_gtsrb_test_dataset('dataset', 'GTSRB_Test.zip')
```

# Preparation of GTSRB training and test datasets

```python
import zipfile
import os
import shutil
```

```python
training_dataset_folder = 'dataset'
training_dataset_zip_file = 'GTSRB_Training.zip'
test_dataset_folder = 'dataset'
test_dataset_zip_file = 'GTSRB_Test.zip'

def prepare_dataset(training_zip_filepath, test_zip_filepath, unzip_to_folder):

    # Clean up 'dataset' folder:
    # delete 'dataset/train_images' if directory exists
    if os.path.exists(os.path.join(unzip_to_folder, 'train_images')): 
        shutil.rmtree(os.path.join(unzip_to_folder, 'train_images'))
        
    # Clean up 'dataset' folder:
    # delete 'dataset/test_images' if directory exists    
    if os.path.exists(os.path.join(unzip_to_folder, 'test_images')): 
        shutil.rmtree(os.path.join(unzip_to_folder, 'test_images'))

    # Clean up 'dataset' folder:
    # delete 'dataset/val_images' if directory exists
    if os.path.exists(os.path.join(unzip_to_folder, 'val_images')): 
        shutil.rmtree(os.path.join(unzip_to_folder, 'val_images'))
    
    # Unzip GTSRB Training dataset
    print('BEGIN to unzip file ' + training_zip_filepath + ' to folder ' + unzip_to_folder)
    
    zip_ref = zipfile.ZipFile(training_zip_filepath, 'r')
    zip_ref.extractall(unzip_to_folder)
    zip_ref.close()
    
    print('END of unzipping file ' + training_zip_filepath + ' to folder ' + unzip_to_folder + '\n')
    
    
    shutil.move(os.path.join(unzip_to_folder, 'GTSRB/Final_Training/Images'), 
                os.path.join(unzip_to_folder, 'train_images'))
    
    
    # Unzip GTSRB Test dataset
    print('BEGIN to unzip file ' + test_zip_filepath + ' to folder ' + unzip_to_folder)
    
    zip_ref = zipfile.ZipFile(test_zip_filepath, 'r')
    zip_ref.extractall(unzip_to_folder)
    zip_ref.close()
    
    print('END of unzipping file ' + test_zip_filepath + ' to folder ' + unzip_to_folder + '\n')
    
    
    shutil.move(os.path.join(unzip_to_folder, 'GTSRB/Final_Test/Images'), 
                os.path.join(unzip_to_folder, 'test_images'))
    shutil.rmtree(os.path.join(unzip_to_folder, 'GTSRB'))
    
    
    # Make validation_data in directory 'dataset' by using images 00000*, 00001* and 00002* of each class
    print('BEGIN making validation_data')
    
    train_folder = os.path.join(unzip_to_folder, 'train_images')
    val_folder = unzip_to_folder + '/val_images'
    if not os.path.isdir(val_folder):
        os.mkdir(val_folder)
        
        for dirs in os.listdir(train_folder):
            if dirs.startswith('000'):
                os.mkdir(val_folder + '/' + dirs)
                for f in os.listdir(train_folder + '/' + dirs):
                    if f.startswith('00000') or f.startswith('00001') or f.startswith('00002'):
                        # Move file to validation folder
                        os.rename(train_folder + '/' + dirs + '/' + f, val_folder + '/' + dirs + '/' + f)
                        
    print('END making validation_data.\n')
    print('END of preparing datasets.\n')
    
```

```python
# Unzip GTSRB Training and Test Datasets and create Validation Dataset
prepare_dataset(os.path.join(training_dataset_folder, training_dataset_zip_file), 
                os.path.join(test_dataset_folder, test_dataset_zip_file),
                'dataset')
```

```python
# Create directory 'data' if it does not exist
if not os.path.exists('data'):
    os.mkdir('data')

ppm_images_path = 'dataset/test_images/'
png_images_path = 'data/png_test_images/'


if not os.path.exists(ppm_images_path):
    print("The specified folder '" + ppm_images_path + 
          "' does not exist.\nPlease download and prepare GTSRB dataset first!")
    
# Create directory 'data/png_test_images'
if not os.path.exists(png_images_path):
    os.mkdir(png_images_path)
else:
    # Clean up directory 'data/png_test_images'
    shutil.rmtree(png_images_path)
    os.mkdir(png_images_path)

    
def convert_ppm_to_png(source_path, destination_path):
    # Converts ppm-images (source_path) to png-images (destination_path)
    
    print('BEGIN conversion: PPM-Image files to PNG-Image files.')
    
    # Create directory 'data/png_test_images/' if it does not exist
    if not os.path.exists(destination_path):
        os.makedirs(destination_path)
    
    # Load each .ppm file of 'dataset/test_images/' and save each image as .png file
    for file in os.listdir(source_path):
        if file.endswith('.ppm'):
            im = Image.open(os.path.join(source_path, file))
            save_path = destination_path + os.path.splitext(file)[0] + '.png'
            im.save(save_path)
            
    print('END conversion: PPM-Image files to PNG-Image files.\n')

    
convert_ppm_to_png(ppm_images_path, png_images_path)
```

```python
def load_image(file_path):
    # Open image from file_path and store it as numpy array
    im = PIL.Image.open(file_path)
    im = np.asarray(im)
    return im - 127.5



def predict_single_image(model, img):
    # Make prediction on single image
    imgArr = (np.expand_dims(img,0))
    imgArr = imgArr/255
    return model.predict(imgArr)[0],img



def make_prediction(input_folder, output_folder, log_path, shape, model, 
                    min_confidence, max_confidence):
    # Make prediction of each png-image 'data/png_test_images/' together with loaded model,
    # and store png-image in 'data/predicted_png_test_images/', if it is within confidence range
    
    print('BEGIN with prediction of PNG-Images on Aphrodite Model.\n')
    
    if not os.path.exists('data'):
        print("The specified folder 'data' does not exist." +
              "\nPlease Convert PPM-Image files to PNG-Image-Files first!")

    if not os.path.exists(input_folder):
        print("The specified folder '" + input_folder + "' does not exist." +
              "\nPlease Convert PPM-Image files to PNG-Image-Files first!")        
        
    # Create directory 'data/predicted_png_test_images/' if it does not exist
    if not os.path.exists(output_folder):
        os.mkdir(output_folder)
    else:
        # Clean up directory 'data/predicted_png_test_images/'
        shutil.rmtree(output_folder)
        os.mkdir(output_folder)
        
        
    # Create directory 'logs/', if it does not exist
    if not os.path.exists(log_path):
        os.mkdir(log_path)
    else:
        # Clean up directory 'logs/'
        shutil.rmtree(log_path)
        os.mkdir(log_path)


    # Create Log file 'logs/prediction_100-100_Aphrodite.txt'
    log_filepath = log_path + 'prediction_' + confidence_range + '_Aphrodite.txt'
    open(log_filepath, 'w').close()
    
    
    i = 0
    j = 0
    
    # Load, resize, and make prediction on each png-image of directory 'data/png_test_images/'
    for file in os.listdir(input_folder):
        
        i += 1
        
        if i % 1000 == 0:
            print('Process File: ' + str(i) + '/' + str(len(os.listdir(input_folder))))
        
        if file.endswith('.png'):
            filepath = os.path.join(input_folder, file)

            # load png image from directory 'data/png_test_images/'
            loaded_png_image = load_image(filepath)

            # resize loaded_png_image to 64 x 64
            resized_png_image = resize(loaded_png_image, (shape, shape))

            # Make prediciton on every resized png image and create prediction-log
            x = np.expand_dims(resized_png_image, axis=0)
            preds = model.predict(x)
            label = np.argmax(preds)
            image_score = predict_single_image(model, resized_png_image)[0]
            image_label_score = image_score[label]

            pred_result = (filepath + '  --  Score=' + str(image_label_score))

            # If prediction on png-image is within confidence range (min_confidence and max_confidence),
            # then append result to 'logs/prediction_100-100_Aphrodite.txt'
            if float(image_label_score) >= min_confidence/100 and float(image_label_score) <= max_confidence/100:

                j +=1
                
                if j % 100 == 0:
                    print('Number of Files within confidence interval: ' + str(j))
                
                # Create directory 'data/predicted_png_test_images/100-100_Aphrodite' if it does not exist
                if not os.path.exists(dest_path_subdir):
                    os.makedirs(dest_path_subdir)

                # Append results to 'logs/prediction_100-100_Aphrodite.txt'
                with open(log_filepath, 'a') as myfile:
                    myfile.write(str(pred_result)+'\n')

                # Filepath: 'data/predicted_png_test_images/100-100_Aphrodite/<PNG-IMAGE>'
                outputfile = os.path.join(dest_path_subdir, file)

                # Copy Input image from 'data/png_test_images/' 
                # to 'data/predicted_png_test_images/100-100_Aphrodite/<PNG-IMAGE>'
                copyfile(filepath, outputfile)
                
    print('Process File: ' + str(i) + '/' + str(len(os.listdir(input_folder))))
    print('Number of Files within confidence interval: ' + str(j))
    print('\nEND of prediction.\n')
    
```

```python

def show_image(image, grayscale = True, ax=None, title=''):
    # Show image as plot
    if ax is None:
        plt.figure()
    plt.axis('off')
    
    if len(image.shape) == 2 or grayscale == True:
        if len(image.shape) == 3:
            image = np.sum(np.abs(image), axis=2)
            
        vmax = np.percentile(image, 99)
        vmin = np.min(image)

        plt.imshow(image, cmap=plt.cm.gray, vmin=vmin, vmax=vmax)
        plt.title(title)
        
    else:
        image = image + 127.5
        image = image.astype('uint8')
        
        plt.imshow(image)
        plt.title(title)
        
        
def load_image(file_path):
    # Open image from file_path and store it as numpy array
    im = PIL.Image.open(file_path)
    im = np.asarray(im)
    return im - 127.5
        
        
def auxiliary_function(mask, destination_filename_path):
    # Store each plotted image in sub directory of 'saliency_map_data/'
    plot.rcParams['figure.figsize'][0] = 0.80 # Set output size to 64 x 64 pixels
    plot.rcParams['figure.figsize'][1] = 0.80 # Set output size to 64 x 64 pixels
    plt.figure(figsize=(10,10))
    show_image(mask) # Saliency Map Creation
    plot.gca().set_axis_off()
    plot.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, hspace = 0, wspace = 0)
    plot.margins(0,0)
    plot.gca().xaxis.set_major_locator(plot.NullLocator())
    plot.gca().yaxis.set_major_locator(plot.NullLocator())
    plot.rcParams['figure.figsize'][0] = 0.80 # Set output size to 64 x 64 pixels
    plot.rcParams['figure.figsize'][1] = 0.80 # Set output size to 64 x 64 pixels
    # Save computed image with shape 64 x 64 pixels
    plot.savefig(destination_filename_path, bbox_inches='tight', pad_inches=0)
 
        
def create_vanilla_saliency(model, resized_png_image, destination_filename_path):
    # Creates a vanilla saliency map
    vanilla = GradientSaliency(model)
    mask = vanilla.get_mask(resized_png_image)
    auxiliary_function(mask, destination_filename_path)

    
def create_smoothed_vanilla_saliency(model, resized_png_image, destination_filename_path):
    # Creates a smoothed vanilla saliency map
    vanilla = GradientSaliency(model)
    mask = vanilla.get_smoothed_mask(resized_png_image)
    auxiliary_function(mask, destination_filename_path)
    
    
def create_guided_backpropagation_saliency(model, resized_png_image, destination_filename_path):
    # Applies guided backpropagation to generate saliency map of input image
    guided_bprop = GuidedBackprop(model)
    mask = guided_bprop.get_mask(resized_png_image)
    auxiliary_function(mask, destination_filename_path)
    
    
def create_smoothed_guided_backpropagation_saliency(model, resized_png_image, destination_filename_path):
    # Applies smoothed guided backpropagation to generate saliency map of input image
    x = np.expand_dims(resized_png_image, axis=0)
    guided_bprop = GuidedBackprop(model)
    mask = guided_bprop.get_smoothed_mask(x[0])
    auxiliary_function(mask, destination_filename_path)
    
    
def create_integrated_gradient_saliency(model, resized_png_image, destination_filename_path):
    # Applies integrated gradient to generate saliency map of input image
    x = np.expand_dims(resized_png_image, axis=0)
    inter_grad = IntegratedGradients(model)
    mask = inter_grad.get_mask(x[0])
    auxiliary_function(mask, destination_filename_path)
    
    
def create_smoothed_integrated_gradient_saliency(model, resized_png_image, destination_filename_path):
    # Applies smoothed integrated gradient to generate saliency map of input image
    x = np.expand_dims(resized_png_image, axis=0)
    inter_grad = IntegratedGradients(model)
    mask = inter_grad.get_smoothed_mask(x[0])
    auxiliary_function(mask, destination_filename_path)
    
```

```python
def create_fs_structure(file, saliency_map_path, method):
    # Auxiliary function to create sub directory structure of 'saliency_map_data', as well as filenames
        
    destination_path = os.path.join(saliency_map_path, method)
    destination_filename = method + '_' + os.path.splitext(file)[0] + '.png'
    
    # Example path:
    # 'saliency_map_data/Smoothed-Guided-Backprop-Saliency_Aphrodite/Smoothed-Guided-Backprop-Saliency_04577.png'
    destination_filename_path = os.path.join(destination_path, destination_filename)

    # Create directory 'destination_path' if it does not exist
    # i.e. 'saliency_map_data/Smoothed-Guided-Backprop-Saliency_Aphrodite/''
    if not os.path.exists(destination_path):
        os.makedirs(destination_path)
        
    return destination_filename_path


def process_png_files(input_folder, shape, model, _van, _sm_van, _g_backp, _sm_g_backp, _int_grad, _sm_int_grad):

    print('BEGIN Saliency Map Creation.\n')
    
    # Directory where all created images (based on Saliency Maps creation) are stored
    saliency_map_path = 'saliency_map_data'
    
    # Create directory 'saliency_map_data' if it does not exist
    if not os.path.exists(saliency_map_path):
        os.mkdir(saliency_map_path)
    
    # Number of files counter
    i = 0
    
    for file in os.listdir(input_folder):
        
        i += 1
        
        if i % 10 == 0:
            print('Process File: ' + str(i) + '/' + str(len(os.listdir(input_folder))))
        
        if file.endswith('.png'):
            filepath = os.path.join(input_folder, file)
            
            # Load png image from file system
            loaded_png_image = load_image(filepath)

            # Resize loaded_png_image to shape x shape
            resized_png_image = resize(loaded_png_image, (shape, shape))

            # Apply Vanilla Saliency Map creation on input image(s), if flag is True
            if _van == True:
                method = 'Vanilla-Saliency_Aphrodite'
                destination_filename_path = create_fs_structure(file, saliency_map_path, method)
                create_vanilla_saliency(model, resized_png_image, destination_filename_path)
            
            
            # Apply Smoothed Vanilla Saliency Map creation on input image(s), if flag is True
            if _sm_van == True:
                method = 'Smoothed-Vanilla-Saliency_Aphrodite'
                destination_filename_path = create_fs_structure(file, saliency_map_path, method)
                create_smoothed_vanilla_saliency(model, resized_png_image, destination_filename_path)
                
                
            # Apply Guided Backpropagation Saliency Map creation on input image(s), if flag is True
            if _g_backp == True:
                method = 'Guided-Backprop-Saliency_Aphrodite'
                destination_filename_path = create_fs_structure(file, saliency_map_path, method)
                create_guided_backpropagation_saliency(model, resized_png_image, destination_filename_path)
                
                
            # Apply Smoothed Guided Backpropagation Saliency Map creation on input image(s), if flag is True
            if _sm_g_backp == True:
                method = 'Smoothed-Guided-Backprop-Saliency_Aphrodite'
                destination_filename_path = create_fs_structure(file, saliency_map_path, method)
                create_smoothed_guided_backpropagation_saliency(model, resized_png_image, destination_filename_path)
                
                
            # Apply Integrated Gradient Saliency Map creation on input image(s), if flag is True
            if _int_grad == True:
                method = 'Integr-Gradient-Saliency_Aphrodite'
                destination_filename_path = create_fs_structure(file, saliency_map_path, method)
                create_integrated_gradient_saliency(model, resized_png_image, destination_filename_path)
                
            
            # Apply Smoothed Integrated Gradient Saliency Map creation on input image(s), if flag is True
            if _sm_int_grad == True:
                method = 'Smoothed-Integr-Gradient-Saliency_Aphrodite'
                destination_filename_path = create_fs_structure(file, saliency_map_path, method)
                create_smoothed_integrated_gradient_saliency(model, resized_png_image, destination_filename_path)
                
    print('END Saliency Map Creation.\n')
                
```

```python
def transmit_images_from_directory(input_folder, model_filepath, shape, min_confidence):
    # Transmit all images from input directory one after the other to the Remote-Web-Interface
    # Note: Only 60 images per minute can be transmitted to Remote-Web-Interface
    
    # Create directory 'logs' if it does not exist
    if not os.path.exists('logs'):
        os.mkdir('logs')

    # Create or clean up 'logs/Remote-Evaluation-Results_<USED_METHOD>_Aphrodite.txt'
    with open('logs/Remote-Evaluation-Results_' + method + '_' + underlying_model + '.txt', 'w') as myfile:
        myfile.write('DateTime;Filename;Trasi-Class;Trasi-Confidence' + '\n')
            
    # Create or clean up 'logs/Remote-Evaluation-Results-90%+_<USED_METHOD>_Aphrodite.txt'
    with open('logs/Remote-Evaluation-Results-90%+_' + method + '_' + underlying_model + '.txt', 'w') as myfile:
        myfile.write('DateTime;Filename;Trasi-Class;Trasi-Confidence' + '\n')

    counter = 0
        
    for file in os.listdir(input_folder):
        
        # To ensure transmission policy, wait for 61 seconds
        if counter % 60 == 0:
            print('Wait 61 sec for next batch transmission...')
            time.sleep(61)

        filename = os.path.join(input_folder, file)

        file_to_upload = {'image': open(filename,'rb')}

        # Transmit each file with given credentials to URL (Remote-Web-Interface)
        post_request = requests.post(url, files=file_to_upload, data=credentials)

        if post_request.status_code == 200:
            # Status code 200 means that image could be transmitted successfully
            post = post_request.text

            trasi_score = json.loads(post)

            # Extract class info from HTTP-response
            _class = trasi_score[0].get('class')
            
            # Extract confidence to class from HTTP-response
            _confidence = trasi_score[0].get('confidence')
            trasi_resp = filename + ';' + str(_class) + ';' + str(_confidence)
            
            # Write results to 'logs/Remote-Evaluation-Results_<USED_METHOD>_Aphrodite.txt'
            with open('logs/Remote-Evaluation-Results_' + method + '_' + 
                          underlying_model + '.txt', 'a') as myfile:
                myfile.write(str(datetime.datetime.now()) + ';' + trasi_resp + '\n')
            
            # If a confidence level higher than 90 % could be achieved, write results to:
            # 'logs/Remote-Evaluation-Results-90%+_<USED_METHOD>_Aphrodite.txt'
            if _confidence >= 0.90:
                with open('logs/Remote-Evaluation-Results-90%+_' + method + '_' + 
                          underlying_model + '.txt', 'a') as myfile:
                    myfile.write(str(datetime.datetime.now()) + ';' + trasi_resp + '\n')
   
        else:
            # Something went wrong. One reason could be that there was a parallel image transmission,
            # that means there were more than 60 images per minute transmitted to Remote-Web-Interface
            print('HTTP-Statuscode: ' + str(post_request.status_code))
        
        counter += 1
        print('Transmitted file: ' + str(counter) + '/' + str(len(os.listdir(input_folder))))
        
    return trasi_score
```
